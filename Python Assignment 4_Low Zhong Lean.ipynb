```python
import shutil, os

# Ask the user to enter string to search
search_path = input("Enter directory path to search,include \ to the ending of the file path : ")
file_type = "png"


try:
    # Append a directory separator if not already present
    if  (search_path.endswith("/") or search_path.endswith("\\")and search_path[1]==":"and search_path[2]=="\\"):
        os.mkdir(search_path+'\\my_new_copy_folder')

        # Repeat for each file in the directory
        for fname in os.listdir(path=search_path):
             # Apply file type filter
            if fname.endswith(file_type):
                 shutil.copy( search_path+fname, search_path+'\\my_new_copy_folder')


        print("Done, please check "+search_path+'\\my_new_copy_folder')




    else:
        print("Invalid directory path")

    
    
except:
    print("cannot find directory path or the file has existed")
        
```
